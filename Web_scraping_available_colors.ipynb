```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
website = 'https://ackodrive.com/collection/maruti-suzuki-cars/'
```

```python
response = requests.get(website)
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
soup
```

```python
car_listings = soup.find_all("div", class_="styles__Wrapper-sc-cbcba852-0 jtAUjc")
```

```python
car_name = soup.find("h2", class_="styles__ModelTitle-sc-57d31ed8-4 LQrbf").get_text(separator=" ").strip()
```

```python
for car in car_listings:
    # Extract car model name
    model_section = car.find("div", class_="styles__ModelDetails-sc-57d31ed8-0 cXLacK")
    car_name = model_section.find("h2", class_="styles__ModelTitle-sc-57d31ed8-4 LQrbf").get_text(separator=" ").strip()
    
    # Extract variant information
    variant_section = car.find("div", class_="styles__VariantDetails-sc-57d31ed8-11 styles__DesktopVariantDetails-sc-cbcba852-9 kEjWdw")
    variant_name = variant_section.find("h4", class_="styles__VariantTitle-sc-57d31ed8-10 haSBAh").get_text().strip()
    
    # Extract color information
    variant_info = variant_section.find("div", class_="styles__VariantInfo-sc-57d31ed8-12 fuepeG").get_text()
    available_colors_text = variant_info.split("Available in")[-1].strip()

    # Print the car model name, variant name, and available colors
    print(f"{car_name} ({variant_name}): {available_colors_text}")
```
